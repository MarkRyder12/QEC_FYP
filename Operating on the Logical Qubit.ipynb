Operating on the Logical Qubit after it's been encoded into Shor's 9-Qubit Quantum Error Correcting Code

```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt

#Sigma-X   sigmax()
#Sigma-Y   sigmay()
#Sigma-Z   sigmaz()
#Sigma plus   sigmap()
#Sigma minus   sigmam()
#identity   qeye(N)  (N=no. levels in hilbert space)
```

```python
#Logical 0 state encoded onto 9 Qubits
ket0 = Qobj([[0],[1]])
print(ket0)
```

```python
a=basis(3, 0) + basis(3, 1) + basis(3, 2)
b=basis(3, 0) + basis(3, 1) + basis(3, 2)
c=tensor(a,b)
print(a)
print(b)
print(c)
d=tensor(c,ket0)
print(d)
```

```python
spin = basis(2,1)
print(spin)
z=sigmaz()
print(z)
print(z * spin)
```

```python
#Logical NOT Gate
ID9x9Matrix = qeye(9)
X = sigmax()
LogicalNOT = tensor(ID9x9Matrix,X)
print(LogicalNOT)
```
